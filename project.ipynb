## Section 3

```python
def _byte_resizer(data):
    return numpy.around(data / 1000000, decimals=2)

data["size_Mb"] = data["size_bytes"].apply(_byte_resizer)
data.drop("size_bytes", axis="columns", inplace=True)
```

```python
plot.subplots(figsize=(10, 8))
bins = [0.00, 10.00, 20.00, 50.00, 100.00, 200.00, 500.00, 1000.00, 2000.00, numpy.inf]
labels = ["<10m", "10-20m", "20-50m", "50-100m", "100-200m", "200-500m", "500-1000m", "1-2G", ">2G"]

freqs = pandas.cut(
	data["size_Mb"],
	bins,
	include_lowest=True,
	labels=labels
)

seaborn.barplot(
	y=freqs.value_counts().values,
	x=freqs.value_counts().index
)
```

```python
bins = [-numpy.inf, 0.00, numpy.inf]
labels = ["free", "paid"]
colors = ['lightcoral', 'yellowgreen']

data["price_categories"] = pandas.cut(
	data["price"], 
	bins,
	include_lowest=True,
	labels=labels
)
```

```python
fig, axis = plot.subplots(figsize=(10, 5))
price_data = data["price_categories"].value_counts()

plot.pie(
	price_data.values,
	labels=labels,
	colors=colors,
	autopct='%1.1f%%',
	shadow=True
)

centre_circle = plot.Circle(
	(0,0),0.75,
	color='black',
	fc='white',
	linewidth=1.25
)

fig = plot.gcf()
fig.gca().add_artist(centre_circle)

plot.axis('equal')
```

```python
free_apps = data.loc[data["price_categories"] == "free"]
paid_apps = data.loc[data["price_categories"] == "paid"]

free_apps_rated = free_apps.sort_values(by=["rating_count_tot"], ascending=False)
paid_apps_rated = paid_apps.sort_values(by=["rating_count_tot"], ascending=False)
```

```python
seaborn.barplot(x=free_apps_rated["rating_count_tot"][:10], y=free_apps_rated["track_name"][:10])
```

```python
seaborn.barplot(x=paid_apps_rated["rating_count_tot"][:10], y=paid_apps_rated["track_name"][:10])
```

## Section 4

```python
genres = data["prime_genre"].value_counts()
genres.sort_values(ascending=False, inplace=True)

plot.subplots(figsize=(20, 10))
seaborn.barplot(x=genres.values, y=genres.index, order=genres.index, orient="h")
```

```python
game_filter = data["prime_genre"] == "Games"
games = data[game_filter]
games.head()
```

```python
prices = (games["price"].value_counts()) / (games["price"].shape[0]) * 100
prices.sort_values(ascending=False, inplace=True)

plot.subplots(figsize=(20, 10))
axis = seaborn.barplot(y=prices.values, x=prices.index, order=prices.index)
axis.set(xlabel="USD", ylabel="percent (%)")
```

### More Visualizations

```python
free_games = games.loc[games["price_categories"] == "free"]
paid_games = games.loc[games["price_categories"] == "paid"]

free_games_rated = free_games.sort_values(by=["rating_count_tot"], ascending=False)
paid_games_rated = paid_games.sort_values(by=["rating_count_tot"], ascending=False)
```

```python
fig = plot.figure(figsize=(20, 10))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

seaborn.barplot(x=free_games_rated["rating_count_tot"][:10], y=free_games_rated["track_name"][:10], ax=ax1)
seaborn.barplot(x=paid_games_rated["rating_count_tot"][:10], y=paid_games_rated["track_name"][:10], ax=ax2)
```
